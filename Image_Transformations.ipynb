```python
#Importing the necessary libraries
import cv2
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Scaling
img = cv2.imread('/content/angel.jpg')
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_NEAREST)
plt.imshow(img)
```

```python
plt.imshow(res)
```

```python
# Translation
rows,cols = img.shape[:2]
M = np.float32([[1,0,500],[0,1,400]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
# Rotation
M = cv2.getRotationMatrix2D((cols/2,rows/2),50,1)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst,'gray')
```

```python
# Morphological Transformations of Images
# Erosion
kernel = np.ones((3,3),np.uint8)
plt.imshow(img,'gray')
```

```python
erosion = cv2.erode(img,kernel,iterations = 1)
plt.imshow(erosion,'gray')
```

```python
# Dilation
dilation = cv2.dilate(img,kernel,iterations = 1)
plt.imshow(dilation,'gray')
```

```python
# Opening
img=cv2.imread('/content/angel.jpg')[0:600,0:1000]
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

```python
# Closing
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
```
